```python
import pandas as pd

# filepath_einkommen = 'estv_income_rates_einkommen.csv'
# filepath_vermoegen = 'estv_income_rates_vermoegen.csv'
filepath_steuern = 'estv_income_rates.csv'
filepath_gemeinden = 'gemeindeverzeichnis_schweiz.csv'
filepath_locationfile ='../TablesDB/Location_v3_enriched.csv'

# columns_einkommen = ['Kantons-Id','Kanton','BfS-Id','Gemeinde','Kantonssteuer','Gemeindesteuer']
# columns_vermoegen = ['Kantons-Id','Kanton','BfS-Id','Gemeinde','Kantonssteuer','Gemeindesteuer']
columns_steuern = ['Kantons-Id','Kanton','BfS-Id','Gemeinde','Einkommenssteuer_Kanton','Einkommenssteuer_Gemeinde','Vermögenssteuer_Kanton','Vermögenssteuer_Gemeinde']
columns_gemeinden = ['Ortschaftsname','PLZ','Gemeindename','BFS-Nr','Kantonskürzel']

# df_einkommen_st = pd.read_csv(filepath_einkommen)[columns_einkommen]
# df_vermoegen_st = pd.read_csv(filepath_vermoegen)[columns_vermoegen]
df_steuern = pd.read_csv(filepath_steuern)[columns_steuern]
df_gemeinden = pd.read_csv(filepath_gemeinden, sep=";")[columns_gemeinden]
df_locationfile = pd.read_csv(filepath_locationfile)

df_gemeinden.rename(columns={'BFS-Nr':'BfS-Id', 'PLZ':'ZIP'}, inplace=True)

df_steuern



```

<h3>Converting the ZIP values in the Location File from float to int<h3/>

```python
df_locationfile['ZIP'] = pd.to_numeric(df_locationfile['ZIP'], errors='coerce').astype('Int64')

df_locationfile.info()

```

<h3>Merging the dataframes<h3/>

```python
df_merged = pd.merge(df_locationfile, df_gemeinden, on='ZIP',how='left')
df_merged = pd.merge(df_merged, df_steuern, on='BfS-Id',how='left')
df_merged.head()
```
