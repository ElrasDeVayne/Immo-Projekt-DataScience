<h1>opendata.swiss API Notebook</h1>

<p>This notebook is dedicated to call the opendata.swiss API and extract data for our ADS Project. The data will enrich our rental price prediction models with relevant statistical data from Canton of Zurich. Most notably we are interested in the tax rates ("Steuerfuss") per municipality.<p> 

```python
import requests
import pandas as pd
import io

# CKAN API endpoint URL
api_url = "https://ckan.opendata.swiss/api/3/action/resource_show"

# Resource ID for the desired resource
resource_id = "c792c06b-9d43-4f63-bac3-f2e89006194d"  # Example resource ID

# Parameters for the API request
params = {
    "id": resource_id
}

# Send a GET request to retrieve resource information
response = requests.get(api_url, params=params)

# Check if the request was successful (status code 200)
if response.status_code == 200:
    # Parse the JSON response
    resource_info = response.json()

    # Get the URL of the resource
    resource_url = resource_info['result']['url']

    # Download the data from the resource URL
    data_response = requests.get(resource_url)

    # Load the data into a pandas DataFrame
    dataset_df = pd.read_csv(io.BytesIO(data_response.content))

   
else:
    print("Failed to retrieve resource information. Status code:", response.status_code)
```
