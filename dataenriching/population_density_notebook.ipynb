## Load the data

```python
import pandas as pd

filepath = 'population_switzerland.csv'

df_population = pd.read_csv(filepath)
df_population.head(1)
```

```python
filepath = '../TablesDB/Location_v3_enriched_v3.csv'

df_locationfile = pd.read_csv(filepath)
df_locationfile.head(1)
```

<a id="2"></a> <br>
## Merge the two dataframes on the BFS number

```python
data_merged = pd.merge(df_locationfile, df_population[['Gemeindecode','Bevölkerungs-dichte pro km²']], left_on='BFS_NUMMER', right_on='Gemeindecode',how='left')
data_merged.head()
```

```python
data_merged.rename(columns={'Bevölkerungs-dichte pro km²':'population_density'}, inplace=True)
```

<a id="3"></a> <br>
## Safe to csv

```python
filepath = '../TablesDB/Location_v3_enriched_v4.csv'
data_merged.to_csv(filepath, index=False)
```
