```python
import pandas as pd
# Load the DataFrame 
df = pd.read_csv('../TablesDB/Location_v3_enriched_v5.csv')
df.head()
```

```python
selected_columns = ['LocationId','Street','ZIP','longitude','latitude','#supermarkets','#foodandbeverage','BFS_NUMMER','NAME','Einkommenssteuer_Kanton','Einkommenssteuer_Gemeinde','Vermögenssteuer_Kanton','Vermögenssteuer_Gemeinde','population_density','#public_transport_stops']

df = df[selected_columns]
df.head()
```

```python
column_mapping = {'LocationId':'LocationId',
                  'Street':'street',
                  'ZIP':'zip',
                  'longitude':'longitude',
                  'latitude':'latitude',
                  '#supermarkets':'supermarket_count',
                  '#foodandbeverage':'foodandbeverage_count',
                  'BFS_NUMMER':'bfs_number',
                  'NAME':'municipality',
                  'Einkommenssteuer_Kanton':'incometax_canton',
                  'Einkommenssteuer_Gemeinde':'incometax_municipality',
                  'Vermögenssteuer_Kanton':'wealthtax_canton',
                  'Vermögenssteuer_Gemeinde':'wealthtax_municipality',
                  'population_density':'population_density',
                  '#public_transport_stops':'public_transport_count'}

df.rename(columns=column_mapping, inplace=True)

df.head()
```

```python
#reorder columns
desired_order = ['LocationId',
                'street',
                'zip',
                'latitude',
                'longitude',
                'bfs_number',
                'municipality',
                'incometax_canton',
                'incometax_municipality',
                'wealthtax_canton',
                'wealthtax_municipality',
                'population_density',
                'public_transport_count',
                'supermarket_count',
                'foodandbeverage_count']

df = df[desired_order]

df.head(1)
```

```python
filepath = '../TablesDB/Location_v3_enriched_v6.csv'
df.to_csv(filepath, index=False)
```
